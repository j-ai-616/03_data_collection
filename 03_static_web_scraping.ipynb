# 정적 웹 페이지 스크래핑

## Web Crawling vs Web Scraping
- Web Crawling: 여러 웹페이지를 자동으로 순회하며 페이지를 수집하는 작업
- Web Scraping: 웹 페이지의 HTML을 분석해서 필요한 데이터만 추출하는 작업

## 정적 웹 페이지 vs 동적 웹 페이지
- 정적 웹 페이지(SSR: Server Side Rendering)
- 동적 웹 페이지(CSR: Client Side Rendering)

## 웹 서비스 흐름
1. 클라이언트 http://example.com 요청
2. 웹 서버 HTML 문서 응답
3. 클라이언트 HTML parsing 후 화면 rendering

```python
# 기본 웹 페이지 요청 및 파싱
import requests
from bs4 import BeautifulSoup

url = 'http://www.naver.com'
response = requests.get(url)

html = response.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
# print(type(soup))
print(soup.prettify())
```

## 샘플 페이지를 통해 BeautifulSoup 메소드 동작 확인

### find | findAll

```python
with open('sample.html', 'r', encoding='utf-8') as f:
    html = f.read()
    # HTML 태그를 계층 구조대로 분석해서 가지고 있음
    soup = BeautifulSoup(html, "html.parser")    

    # find_all : 모든 태그 검색
    li_tags = soup.find_all("li")
    for li_tag in li_tags:
        print(li_tag.text)
        
    # find : 첫 번째 태그 검색
    li_tag = soup.find("li")
    print(li_tag)
    
    # soup 객체 뿐만 아니라 반환 된 객체를 대상으로 다시 하위 조회
    a_tag = li_tag.find("a")
    print(a_tag)
    
    # 속성 조회
    url = a_tag.attrs['href']    # dict 형태로 속성 값 저장 되어있음
    url = a_tag.get('href')      # 안전하게 get 메소드로 반환
    print(url)
    
    # 속성을 포함한 조회
    h1_tag = soup.find('h1', {"id": "page-title"})
    print(h1_tag)
    
    # 모든 .section-content 태그 조회
    section_content_tags = soup.find_all("section", {"class": "section-content"})
    print(section_content_tags)
```
