# 동적 웹 스크래핑
- CSR 방식의 페이지는 기존 requests로 요청/응답 하는 것만으로는 실제 내용을 가져올 수 없다.
- Selenium의 웹자동화 패키지를 이용해서 직접 브라우저를 이용해 내용을 가져와야 한다.

## Selenium
- 웹 애플리케이션 테스트 자동화 도구
- 다양한 프로그래밍 언어와 브라우저 지원
- 웹 페이지와 상호작용하고 데이터를 추출하며 사용자의 브라우저 내 행동을 자동화 하는데 유용

## 환경 설정
- selenium : 파이썬 환경에서 selenium 작동
- chromedriver : selenium - chrome 브라우저 브릿지 프로그램 (최근에는 설치 되어 있음)

## selenium 실행

```python
# WebDriver : 브라우저를 자동으로 제어할 수 있게 해주는 도구
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 크롬 브라우저 자동 실행
driver = webdriver.Chrome()

# 웹 페이지 접속 - 브라우저 주소창에 URL 입력하는 것과 동일
driver.get("https://www.naver.com")

# 대기 - 페이지가 로딩 될 때까지 기다리는 시간
# time.sleep(3)

# 종료
# driver.quit()

# webElement 찾기 - 검색창 id 값이 query인 요소를 찾는다.
# search_box = driver.find_element(By.ID, "query")

# webDriverWait를 이용해서 요소가 나타낼 때까지 기다리는 방법으로 개선한다.
search_box = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "query"))
)
print(type(search_box), search_box)

# webElement 조작하기 - 검색창에 "인공지능" 입력 후 엔터키 입력
search_box.send_keys("인공지능")
search_box.send_keys(Keys.ENTER)

# 뉴스 버튼 찾아서 클릭
news_button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((
        By.XPATH,
        "/html/body/div[3]/div[1]/div/div[2]/div[1]/div/div[1]/div/div[1]/div[2]/a"
    ))
)
print(type(news_button), news_button)
news_button.click()

# 뉴스 목록 가져오기
# 뉴스 목록은 여러개의 요소로 구성되어 있으므로 find_elements() 메서드를 사용하러 여러 요소를 한 번에 가져온다.
news_list_tags = driver.find_elements(By.CSS_SELECTOR, "div.fds-news-item-list-tab > div")  # 선택자 활용: #아이디명, .클래스명, 부모 선택자 > 자식 선택자

for i, news_tag in enumerate(news_list_tags):
    print(i, news_tag)
    # 요소 내부의 요소 찾기 - 뉴스 제목, 설명은 news_tag 내부에 존재
    title_tag = news_tag.find_element(By.CSS_SELECTOR, "a[data-heatmap-target='.tit']")
    description_tag = news_tag.find_element(By.CSS_SELECTOR, "a[data-heatmap-target='.body']")
    title = title_tag.text
    description = description_tag.text
    link = title_tag.get_attribute('href')
    print(i, title, description, link, sep = "|")
```
